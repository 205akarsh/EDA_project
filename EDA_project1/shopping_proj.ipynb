```python
# What is the distribution of ages in the dataset?"
import matplotlib.pyplot as plt
data['Age'].plot(kind="hist")
plt.title("Histogram Of Age")
plt.show()
```

```python
# 1.Age Range: The x-axis covers ages from approximately 18 to 70.

# 2.Frequency: The y-axis represents the frequency (count) of individuals in each age group. The bars are relatively similar in height, suggesting no extreme concentration in a particular age group.

```

```python
# What is the gender distribution in the dataset?
data['Gender'].value_counts().plot(kind='bar')
plt.title('Bar Plot Of Gender Distribution')
plt.show()
```

```python
# Which category has the highest average purchase amount?
data.groupby('Category')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
# Which products or items have the highest review ratings?

data.sort_values('Review Rating',ascending= False)
```

```python
# "What is the proportion of different subscription statuses in the dataset?"
data['Subscription Status'].value_counts().plot(kind='pie')
plt.title('Pie chart of subcription')
plt.show()
```

```python
# Is there a relationship between the number of previous purchases and the review rating?
data.head(10).plot(x="Previous Purchases",y='Review Rating',kind='scatter')
plt.show()
```

```python
# Which season has the highest total purchase amount?
data.groupby('Season')['Purchase Amount (USD)'].sum()
```

```python
# "How does the purchase amount vary across different frequencies of purchases?
data.boxplot(column='Purchase Amount (USD)',by='Frequency of Purchases')
plt.title('box plot purchase amkunt based on frequency of purchase')
plt.show()
```

```python
# Which payment method is most preferred by customers?
data['Preferred Payment Method'].value_counts()
```

```python
data['Preferred Payment Method'].value_counts().plot.barh()
plt.title('Payment Methods')
plt.show()
```

```python
# How does the average purchase amount vary by size?
data.groupby('Size')['Purchase Amount (USD)'].mean()
```

```python
# Which payment method generates the highest total purchase amount?
data.groupby('Payment Method')['Purchase Amount (USD)'].sum()
```

```python
data.groupby('Payment Method')['Purchase Amount (USD)'].sum().plot.barh()
plt.title('SUM OF PURCHASE AMOUNT BY PAYMENT METHOD')
plt.show()
```

```python
# Does applying a discount impact the average review rating?
data.groupby('Discount Applied')['Review Rating'].mean()
```

```python
# Which product categories are most common for each color?
data.groupby('Color')["Category"].value_counts()
```

```python
data.groupby('Season')["Review Rating"].value_counts()
```

```python
# Does the average review rating vary across different seasons?
data.groupby('Season')["Review Rating"].mean()
```

```python
data.groupby('Season')["Review Rating"].mean().plot.barh()
plt.show()
```

```python
# What is the distribution of products across different categories?
data['Category'].value_counts().plot.pie()
plt.show()
```

```python
# How does the distribution of review ratings differ by gender?
data.boxplot(column='Review Rating',by='Gender')
plt.xlabel('GENDER')
plt.ylabel('REVIEW RATING')
plt.title('REVIEW RATING BY GENDER')
plt.show()
```

```python
# "What is the distribution of purchase amounts?"
data['Purchase Amount (USD)'].plot.hist(bins=10)
plt.show()
```

```python
# Which colors have the highest and lowest average review ratings?
data.groupby('Color')['Review Rating'].mean().head().plot.bar()
plt.xlabel("COLOR")
plt.ylabel('REVIEW RATING')
plt.title('MEAN REVIEW RATING FOR EACH COLOR')
plt.show()
```

```python
# What is the share of total purchase amount for each payment method?
data.groupby('Payment Method')['Purchase Amount (USD)'].sum().plot.pie()
plt.ylabel('')
plt.show()
```
